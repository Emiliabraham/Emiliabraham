Boxplot o también conocida como diagrama de caja srive para ver un resumen de una gran cantidad de datos con medidas descriptivas. También podemos identificar valores atípicos y comparar distribuciones. Además de conocer de una forma cómoda y rápida como el 50% de los valores centrales se distribuyen. Las medidas descriptivas que podemos ver son: 
  - Primer cuartil: el 25% de los valores son menores o igual a este valor 
  - Mediana o segundo cuartil: divide en dos partes iguales la distribución. De forma que  el 50% de los valores son menores o igual a este valor 
  - Tercer cuartil: el 75% de los valores son menores o igual a este valor 
  - Rango intercuartílico: diferencia entre el valor del tercer cuartil y el primer cuartil
